```python
df= df.replace({'OrderDate':'DataPedido', 'ShipDate':'DataEnvio', 'UnitPrice':'PreçoUnitario', 'UnitPriceDiscountPct':'PctDescUnit', 'SalesAmount':'TotalVendas', 'TaxAmt':'Imposto', 'Freight':'Frete'})
```

```python
df
```

```python
df['lucro'] = df['UnitPrice'] - (df['TaxAmt'] + df['Freight'])
```

```python
df.head(1)
```

```python
#total lucro
round(df['lucro'].sum(),2)
```

```python
df['TempoEnvio'] = (df['ShipDate'] - df['OrderDate']).dt.days
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.groupby(df['OrderDate'].dt.year)['lucro'].sum()
```

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
#lucro anual
lucro_ano = df.groupby(df['OrderDate'].dt.year)['lucro'].sum().reset_index()
lucro_ano
```

```python
#Produtos com maiores lucros
df.groupby("ProductKey")['lucro'].sum().sort_values(ascending=False)
```

```python
df.groupby(df['OrderDate'].dt.year)['lucro'].sum().plot.bar(title='Lucro anual')
plt.xlabel('Ano')
plt.ylabel('Lucro');
```

```python
df_2018 = df[df['OrderDate'].dt.year == 2018]
df_2018.head(2)
```

```python
df_2018.groupby(df_2018['OrderDate'].dt.month)['lucro'].sum().plot(title="Lucro mensal de 2018")
plt.xlabel("Ano")
plt.ylabel("Lucro");
```
